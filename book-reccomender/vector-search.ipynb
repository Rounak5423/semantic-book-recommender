```python
books[["tagged_description"]].to_csv("tagged_description.txt",index=False,header=False)
```

```python
raw_documents = TextLoader("tagged_description.txt",encoding="utf-8").load()
text_splitter= CharacterTextSplitter(chunk_size=10000,chunk_overlap=0,separator="\n")
documents=text_splitter.split_documents(raw_documents)
```

```python
documents[0]
```

```python
from langchain_google_genai import GoogleGenerativeAIEmbeddings
import torch
```

```python
from langchain_community.vectorstores import Chroma
from langchain_community.embeddings import HuggingFaceEmbeddings


embedding = HuggingFaceEmbeddings(
    model_name="all-MiniLM-L6-v2"
)

db_books = Chroma.from_documents(
    documents,
    embedding=embedding
)
```

```python
query= "A book to teach children about nature"
docs =db_books.similarity_search(query,k=10)
docs
```

```python
import re

books[books["isbn13"] == int(re.findall(r"\d{10,13}", docs[0].page_content)[0])]
```

```python
def retrieve_semantic_recommendations(
        query: str,
        top_k: int = 10,

) -> pd.DataFrame:
    recs=db_books.similarity_search(query,k=5)

    book_list =[]
    for i in range(0,len(recs)):
         book_list += [int(re.findall(r"\d{10,13}", recs[i].page_content)[0])]


    return books[books["isbn13"].isin(book_list)].head(top_k)
```

```python
retrieve_semantic_recommendations("A book to teach children about nature")
```
